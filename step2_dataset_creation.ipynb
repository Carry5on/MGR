Zmiana kodowania w plikach na utf-8 

```python
import chardet

def detect_encoding(file_path):
    with open(file_path, "rb") as f:
        raw_data = f.read(100000)  # Próbka pierwszych 100 KB
        result = chardet.detect(raw_data)
        return result['encoding']
```

```python
import os

folder_path = "dane/main data/klimat 2021-2023/unzipped"
output_folder = "dane/main data/klimat 2021-2023/unzipped/utf8_files/"

os.makedirs(output_folder, exist_ok=True)

for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        input_file = os.path.join(folder_path, file)
        output_file = os.path.join(output_folder, file)

        encoding = detect_encoding(input_file)
        
        if encoding in ["ISO-8859-1", "latin1"]:
            encoding = "cp852" 

        with open(input_file, "r", encoding=encoding, errors="replace") as f:
            content = f.read()

        with open(output_file, "w", encoding="utf-8") as f:
            f.write(content)

        print(f"Przekonwertowano {file} do UTF-8")
```

### Dodanie daty roku kalendarzowego jako kolumny (na bazie daty roku hydrologicznego)

```python
def adjust_year(row):
    if row['Miesiąc'] == 1 or row['Miesiąc'] == 2:  # Jeśli miesiąc to 1 lub 2, zmieniamy rok
        return row['Rok'] - 1
    return row['Rok']  # W przeciwnym razie rok pozostaje bez zmian

# Modyfikacja kolumny 'rok'
hydro_csv_concat['Rok kalendarzowy'] = hydro_csv_concat.apply(adjust_year, axis=1)
```

```python
hydro_csv_concat['Data'] = pd.to_datetime(hydro_csv_concat.rename(columns={'Dzień': 'day', 'Miesiąc kalendarzowy': 'month', 'Rok kalendarzowy': 'year'})\
                                  [['year', 'month', 'day']])
```
