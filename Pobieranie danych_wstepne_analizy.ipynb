wczytanie danych historycznych za 2023 - hydro

```python
import pandas as pd
```

```python
dane2023_mc = pd.read_csv('mies_2023.csv', sep=";",\
            names=['Kod stacji', 'Nazwa stacji', 'Nazwa rzeki/jeziora', 'Rok hydr.',\
                   'Wsk m-ca w r. hydr.', 'Wsk ekstremum', 'Stan wody [cm]', 'Przepływ [m^3/s]',\
                   'Temperatura wody [st. C]', 'Miesiąc kalendarzowy'])
dane2023_mc.head()
```

```python
dane2023_mc
```

Rzeki

Znalezienie stacji z największą liczbą uzupełnionych kolumn potrzebnych do analizy

Odra

```python
kody_stacji_Odra = dane2023_mc[dane2023_mc["Nazwa rzeki/jeziora"] == 'Odra (1)']["Kod stacji"].unique()
```

```python
for kod in kody_stacji_Odra:
    wiersze_dla_kodu = dane2023_mc[dane2023_mc["Kod stacji"] == kod]
    if wiersze_dla_kodu["Stan wody [cm]"].unique()[0] != 99999.999:
        print(kod)
```

Cały zbiór

```python
kody_stacji_all = dane2023_mc["Kod stacji"].unique()
kody_stacji_all.size
```

```python
dane2023_mc["Przepływ [m^3/s]"].unique()
```

```python
for kod in kody_stacji_all:
    wiersze_dla_kodu = dane2023_mc[dane2023_mc["Kod stacji"] == kod]
    if (wiersze_dla_kodu["Stan wody [cm]"].unique()[0]) != 99999.999 \
    and (wiersze_dla_kodu["Temperatura wody [st. C]"].unique()[0]) != 99.9 \
    and (wiersze_dla_kodu["Przepływ [m^3/s]"].unique()[0]) != 99999.999:
        print(kod)
```

```python
import matplotlib.pyplot as plt
```
