### Uzupełenienie brakujących wartości w kolumnie Suma opadów na podstawie informacji z kolumny 
### Rodzaj opadu [S/W/]

```python
# Sortowanie danych
data = data.sort_values(by=['Nazwa stacji', 'Data'])

# Nowa kolumna do wyników
data['Suma dobowa opadów nowa [mm]'] = data['Suma dobowa opadów [mm]']

def rozdziel_opady(grupa):
    grupa = grupa.copy()
    i = 0
    while i < len(grupa):
        if grupa.iloc[i]['Suma dobowa opadów [mm]'] == 0.0 and pd.notna(grupa.iloc[i]['Rodzaj opadu [S/W/]']):
            start = i
            # Szukamy końca serii zerowej sumy opadów z Rodzajem opadu
            while (i < len(grupa) and grupa.iloc[i]['Suma dobowa opadów [mm]'] == 0.0 and
                   pd.notna(grupa.iloc[i]['Rodzaj opadu [S/W/]'])):
                i += 1
            # Jeśli istnieje niezerowa wartość opadów po tej serii, rozdziel opad
            if i < len(grupa) and grupa.iloc[i]['Suma dobowa opadów [mm]'] != 0.0:
                wartosc = grupa.iloc[i]['Suma dobowa opadów [mm]']
                liczba_dni = i - start
                podzial = wartosc / (liczba_dni + 1)
                for j in range(start, i + 1):
                    grupa.at[grupa.index[j], 'Suma dobowa opadów nowa [mm]'] = podzial
        else:
            i += 1
    return grupa

#Zastosowanie funkcji do każdej stacji
data = data.groupby('Nazwa stacji', group_keys=False).apply(rozdziel_opady).reset_index(drop=True)
```
